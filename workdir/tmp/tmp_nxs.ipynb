```python
from Bio.Nexus import Nexus

# Replace 'file.nxs' with the path to your NEXUS file
nexus_file = r"..\data\caffeine.nxs"

# Load the NEXUS file
nexus_data = Nexus.Nexus(nexus_file)

# Access taxa
print("Taxa:", nexus_data.taxlabels)

# Access trees
if nexus_data.trees:
    print("Trees:")
    for tree in nexus_data.trees:
        print(tree)
else:
    print("No trees found in the file.")



```

```python
from Bio import Phylo
import io
from Bio.Nexus import Nexus
import matplotlib.pyplot as plt

# Replace 'file.nxs' with the path to your NEXUS file
nexus_file = r"..\data\caffeine.nxs"
# Load the NEXUS file
nexus_data = Nexus.Nexus(nexus_file)

# Access trees from the NEXUS data
trees = nexus_data.trees

if trees:
    # Convert tree object to Newick string
    newick_tree = trees[0].to_string('newick')
    
    # Use io.StringIO to create a file-like object from the Newick string
    tree_io = io.StringIO(newick_tree)

    # Read the tree from the StringIO object
    tree = Phylo.read(tree_io, 'newick')

    # Draw the tree graphically
    Phylo.draw(tree)
    plt.show()
else:
    print("No trees found in the file.")

```
