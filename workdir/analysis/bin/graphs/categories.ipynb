```python
import matplotlib.pyplot as plt

# Data
vegetation_types = ['Degraded Humid Forest', 'Other Vegetation Types']
coffea_distribution = [90, 10]  # Percentage of Coffea plants
vegetation_coverage = [9, 91]   # Percentage of island coverage

# Plotting
fig, ax = plt.subplots()

# Stacked bar for Coffea distribution
ax.barh(vegetation_types, coffea_distribution, color='lightgreen', label='Coffea Distribution')

# Stacked bar for Vegetation Coverage
ax.barh(vegetation_types, vegetation_coverage, left=coffea_distribution, color='lightgray', label='Vegetation Coverage')

# Add labels and title
ax.set_xlabel('Percentage')
ax.set_title('Coffea Plant Distribution vs Vegetation Coverage in Madagascar')
plt.legend()

# Show plot
plt.show()
```
